# Preprocessing

```python
import pandas as pd

colnames = ['dur','wage1','wage2','wage3','cola','hours','pension','stby_pay',
            'shift_diff','educ_allw','holidays','vacation','lngtrm_disabil',
            'dntl_ins','bereavement','empl_plan','good/bad']
df = pd.read_csv('labor-neg.data',sep=',',names=colnames, header=None,na_values='?')
df
```

```python
df.info()
```

```python
by_mean = ['wage1','wage2','wage3','hours','stby_pay','shift_diff','holidays']
by_mode = ['dur','cola','pension','educ_allw','vacation','lngtrm_disabil',
           'dntl_ins','bereavement','empl_plan']
```

```python
for x in by_mean:
    df[x].fillna(round(df[x].mean()), inplace=True)
    
for x in by_mode:
    df[x].fillna(df[x].mode()[0], inplace=True)
```

```python
df.info()
```

```python
df.to_csv('labor-neg.csv')
```

# Apriori Implementation

```python
df_apr = df.drop(['good/bad'],1)
df_apr.head()
```

```python
from apyori import apriori

records = []
for i in range(0, 39):
    records.append([str(df.values[i,j]) for j in range(0, 16)])
rules = apriori(records, min_support = 0.04, min_confidence = 0.5, min_lift = 3, max_length = 2, target = "rules")
association_results = list(rules)

print(len(association_results))
```

```python
bestSupport = 0
bestConfidence = 0
minimumSupport = 999
minimumConfidence = 999
bestRuleSupport_1, bestRuleSupport_2,bestRuleConfidence_1,bestRuleConfidence_2,minimumRuleConfidence_1,minimumRuleConfidence_2,minimumRuleSupport_1,minimumRuleSupport_2 = [],[],[],[],[],[],[],[]
for item in association_results:
    
    pair = item[0]
    items = [x for x in pair]
    if item[1] >= bestSupport:
        bestRuleSupport_1.append(items[0])
        bestRuleSupport_2.append(items[1])
        bestSupport = item[1]
    if item[2][0][2] >= bestConfidence :
        bestRuleConfidence_1.append(items[0])
        bestRuleConfidence_2.append(items[1])
        bestConfidence = item[2][0][2]
    if item[2][0][2] <= minimumConfidence:
        minimumRuleConfidence_1.append(items[0])
        minimumRuleConfidence_2.append(items[1])
        minimumConfidence = item[2][0][2]
    if item[1] <= minimumSupport:
        minimumRuleSupport_1.append(items[0])
        minimumRuleSupport_2.append(items[1])
        minimumSupport = item[1]                               
     
    print("Rule : ", items[0], " -> " + items[1])
    print("Support : ", str(item[1]))
    print("Confidence : ",str(item[2][0][2]))
    print("Lift : ", str(item[2][0][3]))
    
    print("=============================")
```

```python
print("Maximum dan Minimum Value dengan Contoh Rulenya\n")

print("Maximum Value")  
print("=============================")
print("Rule : ", bestRuleSupport_1[0], " -> " + bestRuleSupport_2[0])
print("Best Support : ", bestSupport)
print("=============================")
print("Rule : ", bestRuleConfidence_1[0], " -> " + bestRuleConfidence_2[0])
print("Best Confidence : ", bestConfidence)
print("=============================\n")
  
print("Minimum Value")  
print("=============================")
print("Rule : ",str(minimumRuleSupport_1[0]), " -> " + str(minimumRuleSupport_2[0]))
print("Minimum Support : ", str(minimumSupport))
print("=============================")
print("Rule : ", str(minimumRuleConfidence_1[0]), " -> " + str(minimumRuleConfidence_2[0]))
print("Minimum Confidence : ", str(minimumConfidence))
print("=============================")
```
